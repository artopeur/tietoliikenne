```python
import numpy as np
import matplotlib.pyplot as plt

taajuustaso = np.zeros(128, dtype=complex)
taajuustaso[4] = complex(1,1)


aikataso = np.fft.ifft(taajuustaso)

plt.figure(1)
plt.subplot(2,2,1)
plt.plot(np.real(aikataso))
plt.subplot(2,2,2)
plt.plot(np.imag(aikataso))
plt.subplot(2,2,3)
plt.plot(np.abs(taajuustaso))

plt.show()
print(taajuustaso)
```
