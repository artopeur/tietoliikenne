```python
import numpy as np
import matplotlib.pyplot as plt

taajuustaso = np.zeros(128, dtype=complex)
taajuustaso[4] = complex(1,1) # lhtetetn QPSK-moduloituna ensimminen
taajuustaso[10] = complex(1,1)
taajuustaso[90] = complex(1,1)

aikataso = np.fft.ifft(taajuustaso)
vastaanotettu = np.fft.fft(aikataso)

plt.figure(1)
plt.subplot(2,2,1)
plt.plot(np.real(aikataso))
plt.subplot(2,2,2)
plt.plot(np.imag(aikataso))
plt.subplot(2,2,3)
plt.stem(np.abs(taajuustaso))
plt.subplot(2,2,4)
plt.stem(np.abs(vastaanotettu))
plt.show()
#print(vastaanotettu)
```
